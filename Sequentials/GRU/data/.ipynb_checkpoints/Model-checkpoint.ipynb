```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import MinMaxScaler
```

```python
inputs = pd.read_csv('X.csv')
inputs.head()
scaler = MinMaxScaler()
inputs[inputs.columns[:-3]] = scaler.fit_transform(inputs[inputs.columns[:-3]])
inputs
```

```python
def sequence_data(X,y,timestep = 5):
    # dims of data [batch,timestep,feature] in our case it's harder cause we need to prepare 5 last matches for both 
    teams_name = pd.read_csv('managers_data.csv').columns[1:]
    teams_results = {}
    teams_data = {}
    final = {}
    Y = {}
    for name in teams_name:
        teams_data[name] = []
        teams_results[name] = []
        final[name] = []
        Y[name] = []
    for idx, match in X.iterrows():
        homeTeam = match[-2]
        awayTeam = match[-1]
        
        teams_data[homeTeam].append(match[:-2])
        teams_data[awayTeam].append(match[:-2])

        
    for team,matches in teams_data.items():
        for idx in range(len(matches)-timestep):
            final[team].append(matches[idx:idx+timestep])
    return final
        
        
            
x_seq = sequence_data(inputs,y)
```

```python
# INPUT = [sample, timesteps, features]
input_data = np.empty((0,5,18))
for key,val in x_seq.items():
    input_data = np.vstack((input_data,np.asarray(val)))
y = []
for idx,sample in enumerate(input_data):
    for jdx,timestep in enumerate(sample):
        if jdx == input_data.shape[1] - 1:
            if timestep[-1] == 'A':
                y.append([1,0,0])
            elif timestep[-1] == 'D':
                y.append([0,1,0])
            elif timestep[-1] == 'H':
                y.append([0,0,1])
                
y = np.asarray(y)
input_data = np.delete(arr = input_data, obj = 17, axis = 2)
print(input_data.shape, y.shape)
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(input_data, y, test_size = 0.3)

np.save('input_data.npy', input_data)
np.save('output_data.npy', y)

'''
Assumptions
- Dropout layer 50%
- Learning rate 1e-4
- Batch size 30 - 50
- Loss MSE
- 

'''

class Model:
    def __init__(self, X, y):
        self.X = X.astype(np.float32)
        self.y = y.astype(np.float32)
        self.model = self.GRU()

        
    def GRU(self, lr = 1e-2):
        model = tf.keras.Sequential()
        model.add(tf.keras.layers.GRU(512,return_sequences=True,input_shape = self.X.shape[1:]))
        model.add(tf.keras.layers.Dropout(0.2))
        
        model.add(tf.keras.layers.GRU(512,return_sequences=False,input_shape = self.X.shape[1:]))
        model.add(tf.keras.layers.Dropout(0.2))
        
        model.add(tf.keras.layers.Dense(3, activation = 'softmax'))
        
        model.compile(optimizer = tf.keras.optimizers.SGD(lr=lr), loss = 'mse', metrics = ['accuracy'])
        
        return model
    
    def train(self, epochs = 50, batch = 150):
        self.history = self.model.fit(self.X, self.y, epochs = epochs)

m = Model(train_X, train_y)
m.train()
```
